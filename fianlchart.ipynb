```python
table = PrettyTable()
```

```python
if isinstance(response, list) and len(response) > 0 and isinstance(response[0], dict):
    
    headers = response[0].keys()
    table.field_names = headers
    
   
    for row in response:
        table.add_row(row.values())
 
    table.border = True
    
   
    print(table)
else:
    print(response)



```
